## Cayley Graphs

A semigroup $S$ is a set with a closed and associative binary operation. One can generate a semigroup by taking elements in a generating set and repeatedly multiplying them. One can make a *right cayley graph*, which is a directed graph, by taking as vertex set the elements of S - factored in terms of the generators - and adjoining vertices $x,y \in V$ if there exists $a$ in the generating set such that $xa = y$. Similarly, one can make the *left cayley graph*, by adjoing vertices $x,y$ if there exists $a$ in the generating set such that $ax = y$. 

```python
from libsemigroups_cppyy import Transformation, Permutation
import libsemigroups_cppyy.froidure_pin as fp
import numpy as np
import networkx
from pyvis.network import Network
```

Define your semigroup creating a Froidure-Pin class, and placing generating elements in the function. The following produces a group (because the generators are permutations). 

```python
#Define your Semigroup. This is the one from the notes.
S = fp.FroidurePin(Permutation([1, 2, 3, 0]), Permutation([1, 0, 2, 3]))
S.size()
```

This is the right cayley graph of S. 

```python
G = fp.right_cayley(S)
G
```

The function right_cayley outputs a networkx graph object. Using networkx.draw we can show it. 

```python
np.random.seed(2)
networkx.draw(G, with_labels = True)
```

This is the left cayley graph of S.

```python
np.random.seed(2)
G = fp.left_cayley(S)
networkx.draw(G, with_labels = True)
```

Using the pyvis functionality, we can play with the nodes of the graph interactively. 

```python
g = Network(notebook=True)
g.toggle_physics(status=False)#physics solver
g.from_nx(G)
g.show("left_cayley.html")
```

# Cayley graphs of transformations

The following semigroups and cayley graph were made in the notes. We construct it here. 

```python
S = fp.FroidurePin(Transformation([1,1,2]), Transformation([1,0,1]))
```

```python
np.random.seed(2)
G = fp.right_cayley(S)
networkx.draw(G, with_labels = True)
```

```python
g = Network(notebook=True)
g.toggle_physics(status=False)#physics solver
g.from_nx(G)
g.show("right_cayley.html")
```

```python
np.random.seed(2)
G = fp.left_cayley(S)
networkx.draw(G, with_labels = True)
```

So left and right cayley graphs of semigroups can look very different!

```python
g = Network(notebook=True)
g.toggle_physics(status=False)#physics solver
g.from_nx(G)
g.show("left_cayley.html")
```
